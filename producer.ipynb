```python
import pandas as pd
from kafka import KafkaConsumer, KafkaProducer
from json import dumps
from time import sleep
import json
```

```python
# create producer object to connect to ec2 instance public IP
producer = KafkaProducer(bootstrap_servers=['13.229.124.127:9092'],
                         value_serializer=lambda x:dumps(x).encode('utf-8')
                        )
```

```python
df = pd.read_csv("sp500_data.csv")
```

```python
# get the latest stock info for each stock
latest_stock_info = df.sort_values('Date').groupby('Ticker').tail(1)
latest_stock_info
```

```python
#simulate real-time stock info feed
while True:
    stock_dict = latest_stock_info.to_dict(orient="records")
    producer.send('jonathan_test', value=stock_dict)
```
